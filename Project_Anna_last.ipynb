```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec
```

```python
#importing web data
df_curr = pd.read_html('https://www.extratime.com/competition/2123/202/2022-world-cup/')
df_curr
```

```python
#merging all DataFrames into one
df_all = pd.DataFrame()
for f in df_curr:
  df_all = pd.concat([df_all,f])

#2 option:
#all_groups = pd.concat([df_1group,df_2group,df_3group,df_4group,df_5group,df_6group,df_7group,df_8group])
#all_groups

df_all
```

```python
#renaming columns because there was a shift in the names during importing from webpage and consolidation of changes
df_all.rename(columns={
    'Unnamed: 0':'Place',
    'Pd':'Country',
    'W':'Played',
    'D':'Won',
    'L':'Draw',
    'GF':'Loss',
    'GA':'Goals For',
    'GD':'Goals Against',
    'Pts':'Goal Difference',
    'Form':'Points',
    'Unnamed: 11':'Form'
}, inplace=True)
df_all
```

```python
#removing columns without data
df_all = df_all.drop(columns = ['Country','Player','Total Goals','Penos'], axis=1)
df_all
```

```python
#reindexing (to do it in place without saving the old index)
df_all.reset_index(inplace=True,drop=True)
df_all
```

```python
#saving data and changes to a file
df_all.to_csv('Fifa_2022.csv')
```

```python
#selecting by columns
df_all[['Team','Points']]
```

```python
#selection of teams that did not lose
df_all[df_all.Loss==0]
```

```python
#selection of teams that received 6 or more points
df_all[df_all.Points>=6]
```

```python
#selection of teams that lost and draw
df_all[(df_all.Draw>=1)&(df_all.Loss>=0)]
```

```python
#select teams that are going to next
df_all_next = df_all[(df_all.Place==1)|(df_all.Place==2)]
df_all_next
```

```python
#median value in specific column
df_all_next['Points']. median ()
```

```python
#sorting teams by points(descending)
df_all_sort = df_all_next.sort_values(by='Points', ascending=False)
df_all_sort
```

```python
#creating graphs of each group by points
fig = plt.figure(figsize=(20, 10))
gs = GridSpec(ncols=4, nrows=2, figure=fig)
ax1 = plt.subplot(gs[0, 0])
ax1.plot(df_1group['Team'], df_1group['Form'],color='red',marker='*')
ax1.set_ylabel('Group A', labelpad = 0, fontsize = 12)
ax1.grid()
ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(df_2group['Team'], df_2group['Form'],marker='*')
ax2.set_ylabel('Group B', labelpad = 0, fontsize = 12)
ax2.grid()
ax3 = fig.add_subplot(gs[0, 2])
ax3.plot(df_3group['Team'], df_3group['Form'],color='red',marker='*')
ax3.set_ylabel('Group C', labelpad = 0, fontsize = 12)
ax3.grid()
ax4 = fig.add_subplot(gs[0, 3])
ax4.plot(df_4group['Team'], df_4group['Form'],marker='*')
ax4.set_ylabel('Group D', labelpad = 0, fontsize = 12)
ax4.grid()
ax5 = plt.subplot(gs[1, 0])
ax5.plot(df_5group['Team'], df_5group['Form'],marker='*')
ax5.set_ylabel('Group E', labelpad = 0, fontsize = 12)
ax5.grid()
ax6 = fig.add_subplot(gs[1, 1])
ax6.plot(df_6group['Team'], df_6group['Form'],color='red',marker='*')
ax6.set_ylabel('Group F', labelpad = 0, fontsize = 12)
ax6.grid()
ax7 = fig.add_subplot(gs[1, 2])
ax7.plot(df_7group['Team'], df_7group['Form'],marker='*')
ax7.set_ylabel('Group G', labelpad = 0, fontsize = 12)
ax7.grid()
ax8 = fig.add_subplot(gs[1, 3])
ax8.plot(df_8group['Team'], df_8group['Form'],color='red',marker='*')
ax8.set_ylabel('Group H', labelpad = 0, fontsize = 12)
ax8.grid()
```

```python
#creating graph of teams that are going to next (compare with median)
fig,ax = plt.subplots(figsize=(17, 7))
ax.plot(df_all_sort['Team'], df_all_sort['Points'], color='green',linewidth=3)
ax.set_ylabel('Points', labelpad = 0, fontsize = 15)
ax.fill_between(df_all_sort['Team'], df_all_sort['Points'], color='skyblue',alpha=0.3)
ax.set_yticks(np.arange(0,10,6))
ax.grid(color='blue',linewidth=1, linestyle='--')
```
